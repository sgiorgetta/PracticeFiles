```python
from pylab import *
```

```python
import numpy as np
```

```python
#let's try the example, a  plot of y = x**2 
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, 'r--')
xlabel('x')
ylabel('y')
title('plot of y = x ** 2')
show()
```

```python
#Let's try a few more examples.  Try dashed lines and the inverse.
```

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y,x,'g*-');
```

```python
#Now let's try object oriented API
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('mytitle');
```

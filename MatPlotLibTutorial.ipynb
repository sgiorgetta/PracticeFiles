# Histogram Example

```python
x = np.random.normal(size=1000)

fig, ax = plt.subplots()

H = ax.hist(x, bins=50, alpha = 0.5, histtype = 'stepfilled')
```

# Pie Plot Example

```python
fracs = [30, 15 ,45, 10]
colors = ['b', 'g', 'r', 'w']

fig, ax = plt.subplots(figsize = (6,6))
pie = ax.pie(fracs, colors=colors, explode=(0,0,0.05,0), shadow=True, labels = ['A', 'B', 'C', 'D'])
```

# Errorbar Plot Example

```python
x= np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots()
plt.errorbar(x, y, dy, fmt = '.k')
```

# Filled Plot Exampled

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where=(y1<y2), color = 'red')
ax.fill_between(x, y1, y2, where=(y1>y2), color = 'blue')
```

# Scatter Plot Example

```python
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50)
s = 500 * np.random.random(50)

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap=plt.cm.jet)

fig.colorbar(im, ax=ax)
im.set_clim(0.0, 1.0)
```

# Contour Plot Example

```python
x = linspace(0, 10, 50)
y = linspace(0, 20, 60)

z = np.cos(y[:,np.newaxis]) * np.sin(x)

fig, ax = plt.subplots()

im = ax.contourf(x, y, z, 100)

im2 = ax.contour(x, y, z, colors = 'k')

fig.colorbar(im, ax=ax)
```

# Showing Images Example 

```python
I = np.random.random((100, 100))

I += np.sin(np.linspace(0, np.pi, 100))

fig, ax = plt.subplots()

im = ax.imshow(I, cmap=plt.cm.jet)

fig.colorbar(im, ax=ax)
```

# 2D Histogram and Hexbin Example 

```python
x, y = np.random.normal(size = (2, 10000))

fig, ax = plt.subplots()
im = ax.hexbin(x, y, gridsize = 20)
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
H = ax.hist2d(x, y, bins = 20)
fig.colorbar(H[3], ax=ax)
```

# Polar Plot Example 

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = 'polar')

theta = np.linspace(0, 10 * np.pi, 1000)
r = np.linspace(0, 10, 1000)

ax.plot(theta, r)
```

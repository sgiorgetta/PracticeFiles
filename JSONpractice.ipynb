```python
# import json
import json
from pandas.io.json import json_normalize
```

```python
# load the json file as string
data = json.load((open('data/world_bank_projects.json')))
```

```python
# normalize extracting the fields needed for the assignment
df = json_normalize(data, 'mjtheme_namecode', ['countryshortname', ['_id', '$oid']] )
```

```python
# check to see if the result looks reasonable
df.head()
```

```python
# get the list of top 10 countries that have the most projects
df.countryshortname.value_counts()
```

```python
# get the top 10 major project themes.  
df.name.value_counts()
```

```python
# Note that there is missing data from the 'name' column invalidating the previous result
# sort the dataframe by code so we can find and fill the empty spots in 'name'
df_sorted = df.sort(['code','name'])
```

```python
#  check to see if we need to do a back or forward fill
df_sorted[df_sorted.code == '1']
```

```python
# replace the "empty" spots, ones with u'', with Nan, then back fill using fillna
df_sorted.replace(u'', np.nan).fillna(method='bfill')
```

```python
# redo the count of top 10 projects to update the numbers after filling in missing data
df_sorted.replace(u'', np.nan).fillna(method='bfill').name.value_counts()
```
